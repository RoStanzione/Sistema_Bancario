```python
def menu():
    menu= '''
[1] = depositar
[2] = sacar
[3] = extrato
[4] = cadastrar novo cliente
[5] = criar nova conta
[6] = listar as contas
[0] = sair
'''
    print(menu)


def depositar(saldo, valor, extrato,/):
    if valor>0:
        saldo= saldo+valor
        extrato+= f'Depósito: R$ {valor:.2f}\n'
        print('Depósito realizado com sucesso!')
    else: print('Digite um valor válido')
    return saldo, extrato


def sacar(*,saldo, valor, extrato, limite_por_saque, numero_saques, limite_saques):

    excedeu_saldo = valor > saldo

    excedeu_limite = valor > limite_por_saque

    excedeu_saques = numero_saques >= limite_saques

    if excedeu_saldo:
        print("Operação falhou! Você não tem saldo suficiente.")

    elif excedeu_limite:
        print("Operação falhou! O valor do saque excede o limite.")

    elif excedeu_saques:
        print("Operação falhou! Número máximo de saques excedido.")

    elif valor > 0:
        saldo -= valor
        extrato += f"Saque: R$ {valor:.2f}\n"
        numero_saques += 1
        print('Saque realizado com sucesso!')

    return (saldo, extrato)

def exibir_extrato(saldo,/,*,extrato):
    print('Extrato'.center(50, '='))
    print(extrato)
    print('Saldo atual: R$', saldo)
    print('Extrato'.center(50, '='))

def criar_usuario(usuarios):
  cpf= input('Digite o cpf do cliente: ')
  usuario= filtrar_usuario(cpf, usuarios)
  if usuario in usuarios:
    print('Usuário já existe')
    print(usuario)
  else: 
    nome= input('Digite o nome do cliente: ')
    data_de_nascimento=input('Digite a data de nascimento (dd-mm-aaaa): ')
    endereco= input('Digite o endereço completo (logradouro- nº- bairro, cidade/sigla do Estado): ')
    usuarios.append({'nome': nome, 'cpf': cpf, 'data de nascimento': data_de_nascimento, 'endereço': endereco})
    print('====Cliente cadastrado com sucesso!====')

def filtrar_usuario(cpf, usuarios):
  for usuario in usuarios:
    if usuario['cpf'] == cpf:
      return(usuario)
    else: print('cliente não cadastrado!')

def criar_conta(agencia, numero_conta, usuarios):
  cpf= input('Digite o cpf do cliente: ')
  usuario=filtrar_usuario(cpf, usuarios)
  if usuario:
    print('====Conta criada com sucesso!====')
    return({'agencia': agencia, 'numero da conta': numero_conta, 'usuario': usuario})
  else: print('----Usuário não encontrado, cadastre o usuário primeiro----')

def listar_contas(contas):
  for conta in contas:
    print(conta)


def conta_bancaria():

    saldo= 500
    extrato= ''
    limite_saque= 500
    limite_saque_diario= 1500
    qtd_saque= 0
    usuarios=[]
    contas=[]
    agencia='0001'


    while True:

        menu()

        opcao= int(input('Digite a opcao desejada: '))



#depositar
        if opcao==1:
            valor= float(input('Digite o valor para depósito: '))
            saldo, extrato=depositar(saldo, valor, extrato)

#sacar
        elif opcao==2:
            valor= float(input('Digite o valor para saque: '))
            saldo, extrato= sacar(
                saldo=saldo,
                valor=valor,
                extrato=extrato,
                limite_por_saque=limite_saque,
                numero_saques=qtd_saque,
                limite_saques=limite_saque_diario
            )

#extrato
        elif opcao==3:
            exibir_extrato(saldo,extrato=extrato)

#cadastrar cliente
        elif opcao==4:
          criar_usuario(usuarios)

#criar conta
        elif opcao==5:
          numero_conta= len(contas)+1
          conta= criar_conta(agencia, numero_conta, usuarios)
          if conta:
            contas.append(conta)

#listar contas
        elif opcao==6:
          listar_contas(contas)


        elif opcao==0:
          break

        else: print('Digite uma opção válida')



```
